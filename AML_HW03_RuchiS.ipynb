### Question 2: t-SNE 

Read this [article](https://erdem.pl/2020/04/t-sne-clearly-explained) on t-SNE, a dimensionality reduction technique for visualization and explain it in your own words in one or two paragraphs.

#### Answer:

Many popular dimensionality reduction techniques (such as PCA) are based on the linear transformations within the feature spaces. But there can be cases where we need to handle non-linear dimensionality reduction. This occurs in cases when the data is not separable by a straight line. One the algorithms that deals with such non linearity is t-SNE, t-distributed Stochastic Neighbor Embedding. Furthermore, t-SNE is iterative and unlike PCA, cannot be applied to another dataset. 

The goal of this algorithm is to take a set of points in a high-dimensional space and find a faithful representation of those points in a lower-dimensional space, typically the 2D plane. The algorithm is non-linear and adapts to the underlying data, performing different transformations on different regions. The algorithm creates a probability distribution that represents similarities between neighbors. Further, SNE performs a binary search for the value of sigma that produces probability distribution with a fixed perplexity that is specified by the user. The article also describes the concept of "Shannon Entropy" and points out that perplexity is positively correlated with the value of mu_i. 

t-SNE then creates a low-dimensional space with the same number of points as in the original space, but points should be randomly spread on a new space. The goal of this algorithm is to find similar probability distribution in low-dimensional space. But here instead of using the Gaussian, we use Student t-distribution with a single degree of freedom. Students-t ‚Äúfalls‚Äù quickly and has a ‚Äúlong tail‚Äù so points won‚Äôt get squashed into a single point. 

To optimize the distribution, t-SNE uses the Kullback-Leibler Divergence. t-SNE performs well itself but there are some improvements that allow it to do even better (Early compression, Early exaggeration). 


### Question 3: PCA

Explain the principle of Principal Component Analysis algorithm and in what sense is the selection of the eigenvectors to represent the data is an optimal choice.  How do you reconstruct (a noisy version of) the original data from the eigenvectors and the loadings?



#### Answer:

Principal Component Analysis (PCA) is a linear dimensionality reduction method in statistics and helps in feature selection of the dataset. But at the same time, it ensures that there is minimum information loss. It does so by creating new uncorrelated variables called the priniple components that successively maximize variance. When we go about finding the principal components that maximizes the variance in our projected data, the approach reduces to solving an eigenvalue/eigenvector of the covariance matrix problem of the mean centered data (x- E(x)) (Mathmematics is beyond the scope of this question). The eigen vectors of the  mean centered data's covariance matrix are seen be the optimal choice when it comes to maximizing the variance of our dimensionally reduced data. 

These eigenvectors of the covariance matrix are actually the directions of the axes that have the greatest variance (principal components) while the eigenvalues are the coefficients attached to the eigenvectors that indicate the amount of variance contained in each principal component.

PCA algorithm in vector form: $ Z = XV$ where X is our original dataset, V is the Eigen vector space matrix and Z is dimentionally reduced data (loading)

**Reconstruction:**

If we want to get our original data back, we can multiply the loadings Z with the transpose of the eigenvector matrix - $\hat{X} = ZV^T$. This would give a noisy version of X as we lost information when we mapped it to a lower dimensional space

### Question 4: Outlier detection using PyOD

Outlier detection, or anomaly detection is usually an unsupervised learning task where the objective is to identify suspicious observations in data. It has been widely used in military surveillance for enemy activities to prevent attacks, intrusion detection in cyber security, fraud detection for credit cards, etc.

PyOD is a comprehensive and scalable Python library for detecting outlying objects in multivariate data. PyOD includes more than 30 detection algorithms and provides unified APIs which makes it quite handy to use. In this question, you will play with PyOD, explore three different outlier detection algorithms and compare their performances. First let's install PyOD.

````python
# install pyod using pip first
# !pip install pyod
````

````python
from pyod.models.knn import KNN
import warnings
warnings.filterwarnings("ignore")
````

````python
import pandas as pd

train_df = pd.read_csv('Q3_train_dataset.csv')
test_df = pd.read_csv('Q3_test_dataset.csv')

X_train = train_df[['X_train_0', 'X_train_1', 'X_train_2', 'X_train_3', 'X_train_4']].to_numpy()
y_train = train_df[['y_train']].to_numpy()
X_test = test_df[['X_test_0', 'X_test_1', 'X_test_2', 'X_test_3', 'X_test_4']].to_numpy()
y_test = test_df[['y_test']].to_numpy()
````

````python
# (a)
from pyod.models.knn import KNN
clf = KNN() # initialize KNN class using the default parameters

# fit the model using X_train
clf.fit(X_train)
y_train_pred = clf.labels_
y_train_scores = clf.decision_scores_
````

````python
from pyod.utils.data import evaluate_print
# predict the outlier labels of X_test using the trained model, compute the raw outlier scores on X_test using decision_function()
# then use evaluate_print() to print out the evaluation results

y_test_pred = clf.predict(X_test)  # outlier labels (0 or 1)
y_test_scores = clf.decision_function(X_test)  # outlier scores

print("\nOn Training Data:")
evaluate_print(clf, y_train, y_train_scores)

print("\nOn Test Data:")
evaluate_print(clf, y_test, y_test_scores)
````

````python
# (b)
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

# Fit pca to X_train and X_test and transform 
train_principalComponents = pca.fit_transform(X_train)
test_principalComponents = pca.transform(X_test)

from pyod.utils.example import visualize
# Visualize the ground truth outliers and predicted outliers using visualize()

visualize(clf, train_principalComponents, y_train, test_principalComponents, 
              y_test, y_train_pred,y_test_pred, show_figure=True, save_figure=False)
````

````python
# (c)
from pyod.models.inne import INNE

clf_inne = INNE()
clf_inne.fit(X_train)
y_train_pred_inne = clf_inne.labels_
y_train_scores_inne= clf_inne.decision_scores_

y_test_pred_inne= clf_inne.predict(X_test)  # outlier labels (0 or 1)
y_test_scores_inne= clf_inne.decision_function(X_test)  # outlier scores
````

````python
print("\nOn Training Data:")
evaluate_print(clf_inne, y_train, y_train_scores_inne)

print("\nOn Test Data:")
evaluate_print(clf_inne, y_test, y_test_scores_inne)
````

````python
visualize(clf_inne, train_principalComponents, y_train, test_principalComponents, 
          y_test, y_train_pred_inne,y_test_pred_inne, show_figure=True, save_figure=False)
````

The ROC for INNE (0.8633) is worse than the ROC (0.9044) for K-Nearest Neighbours. However, when it comes to precision, INNE (0.6) is better than KNN (0.5). This means that INNE is better than KNN at predicting "actual outliers" as "outliers" (higher precision). However in the process, it is also incorrectly predicting "non-outlier points" as "outliers" (lower ROC) when compared to KNN. This is further corroborated in the visuals we are seeing.

e) Calculate 95% confidence intervals for SalePrice with a sample size of 100. 

````python
import scipy.stats as st

np.random.seed(10)
sample_size = 100
sample = np.random.choice(a= data_dm['SalePrice'],
                          size = sample_size)
sample_mean = sample.mean()

#Get the critical Z value
z_critical= st.norm.ppf(.975)

#Get population standard deviation
pop_stdev=data['SalePrice'].std()

#margin of error
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size)) 

#confidence interval
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

#Print confidence interval and true mean value
print(confidence_interval)
print("Population Mean:",data['SalePrice'].mean())
````
